```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
```

```python
#Read the data into a dataframe
adcc_df = pd.read_csv("adcc_historical_data.csv")

#Examine the first couple of rows of the dataframe
####### 
adcc_df.head()
####### 
```

```python
#Print the most popular submissions

submission_counts = adcc_df['submission'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
submission_counts.plot(kind='bar', color='skyblue')

# Adding title and labels
plt.title("Most Popular Submissions")
plt.xlabel("Submission Type")
plt.ylabel("Number of Occurrences")

# Display the bar chart
plt.tight_layout()
plt.grid(axis='y')
plt.show()

```

```python
# Get the best scoring athletes for each year

# Filter the data from 2015 to 2022
filtered_data_set = adcc_df[adcc_df['year'].between(adcc_df['year'].min(), adcc_df['year'].max())]

# Group by year and then get the row with the maximum 'winner_points' for each year after filtering
best_athletes_each_year = filtered_data_set.loc[filtered_data_set.groupby('year')['winner_points'].idxmax()]

# Plotting
plt.figure(figsize=(12, 8))
bars = plt.bar(best_athletes_each_year['year'].astype(str), 
        best_athletes_each_year['winner_points'], 
        color='skyblue', 
        alpha=0.7)

# Label the bars with the best athlete's name
for bar, row in zip(bars, best_athletes_each_year.iterrows()):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, row[1]['winner_name'],
             ha='center', va='bottom', rotation=90, fontsize=9)

# Adding title and labels
plt.title("Best Scoring Athletes Each Year (2015 to " + str(adcc_df['year'].max()) + ")")
plt.xlabel("Year")
plt.ylabel("Points")

# Display the bar chart
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # To ensure that all labels fit well
plt.show()
```

```python
# Fill missing 'submission' values with 'None'
adcc_df['submission'].fillna('None', inplace=True)

print(adcc_df)
```

```python
# Convert 'winner_id' and 'loser_id' to strings as they are not used for calculations
adcc_df['winner_id'] = adcc_df['winner_id'].astype(str)
adcc_df['loser_id'] = adcc_df['loser_id'].astype(str)
```

```python
# Filter the dataframe for matches won by points
points_matches = adcc_df[adcc_df['win_type'] == 'POINTS']

# Calculate summary statistics for 'winner_points' and 'loser_points' in matches won by points
winner_points_stats = points_matches['winner_points'].describe()
print("Winner Points Statistics:")
print(winner_points_stats)

# Display the descriptive statistics for loser points
loser_points_stats = points_matches['loser_points'].describe()
print("\nLoser Points Statistics:")
print(loser_points_stats)
```

```python
The analysis of points in matches decided by scoring reveals a competitive landscape where victories are often achieved with a modest number of points.
The average points scored by winners is 2.57, with a standard deviation of 3.60, indicating a wide range of outcomes from match to match. Notably,
a quarter of the winners won with no points scored, suggesting that penalties against the opponent play a significant role in some victories. 
The median score of 2 points for winners aligns with the competitive nature of the sport, where matches can be closely contested.
For the losers, the average score is slightly negative at -0.19, with a standard deviation of 0.76, reflecting a narrower range of scores. 
This negative average is influenced by the scoring system, where a score of -1 represents matches lost without any points scored, likely due to penalties. 
The median score of 0 for losers indicates that in at least half of the matches, the losing competitors did not score any points,
emphasizing the difficulty of scoring in this highly competitive environment.
    
The maximum scores of 22 points for winners and 6 points for losers highlight the potential for outliers, where matches can see higher scoring under certain circumstances.
However, such instances are exceptions rather than the norm, as indicated by the 75th percentile scores
of 4 for winners and 0 for losers, suggesting that most matches are won with a few decisive moves rather than an accumulation of many points
```

```python
# Filter for matches where the winner's score is -1
matches_with_winner_score_neg1 = adcc_df[adcc_df['winner_points'] == -1]

# Examine the win types for these matches
win_types = matches_with_winner_score_neg1['win_type'].value_counts()

# Check for the presence of submissions in these matches
submissions = matches_with_winner_score_neg1['submission'].value_counts()

# Display the results
print("Win Types for Matches with Winner Score of -1:")
print(win_types)
print("\nSubmissions for Matches with Winner Score of -1:")
print(submissions)
```
